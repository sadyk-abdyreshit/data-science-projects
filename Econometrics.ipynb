```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

data = np.array([50,55,53,60,65,70])
```

### 1st question

```python
data_mean = np.mean(data)
print(data_mean)
```

### 2nd question

```python
plt.plot(data)
# Add a title and labels (optional)
plt.title("Retail Market Sales")
plt.xlabel("Months")
plt.ylabel("Sales (in $1000)")
plt.show
```

### 3rd question

```python
plt.plot(data)
# Add a title and labels (optional)
plt.title("Retail Market Sales")
plt.xlabel("Months")
plt.ylabel("Sales (in $1000)")
plt.show

sales = data

months = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)

model = LinearRegression()
model.fit(months, sales)

next_month = np.array([[7]])
prediction = model.predict(next_month)

intercept = model.intercept_
slope = model.coef_[0]
sales = np.append(sales, prediction)
print(f"Regression Equation: Sales = {intercept:.2f} + {slope:.2f} * Month")
print(f"Predicted Sales for Month 7: {prediction[0]:.2f}")

plt.plot(sales)
plt.show
```

```python
file = pd.read_csv('canada_per_capita_income.csv')
file.head()
```

```python
%matplotlib inline

plt.plot(file.year, file['per capita income (US$)'], c='r')
```

```python
reg = linear_model.LinearRegression()
reg.fit(file[['year']], file[['per capita income (US$)']])
```

```python
next_years = (2017,2018,2019,2020)
next_years = np.array(next_years).reshape(-1,1)
```

```python
predictions = reg.predict(next_years)
```

```python
plt.xlabel('Year', fontsize=12)
plt.ylabel('Per Capita Income (US$)', fontsize=12)
plt.scatter(file.year, file['per capita income (US$)'], color='red', marker='o', label='Actual Data')
plt.plot(next_years, predictions, color='blue', label='Predictions')

# 4. Optional: Draw the regression line through all points
plt.plot(file.year, reg.predict(file[['year']]), color='green', linewidth=1)

plt.legend()
plt.show()
```
