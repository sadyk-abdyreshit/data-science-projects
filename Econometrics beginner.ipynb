```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

data = np.array([50,55,53,60,65,70])
```

### 3rd question

```python
plt.plot(data)
# Add a title and labels (optional)
plt.title("Retail Market Sales")
plt.xlabel("Months")
plt.ylabel("Sales (in $1000)")
plt.show

sales = data

months = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)

model = LinearRegression()
model.fit(months, sales)

next_month = np.array([[7]])
prediction = model.predict(next_month)

intercept = model.intercept_
slope = model.coef_[0]
sales = np.append(sales, prediction)
print(f"Regression Equation: Sales = {intercept:.2f} + {slope:.2f} * Month")
print(f"Predicted Sales for Month 7: {prediction[0]:.2f}")

plt.plot(sales)
plt.show
```

# Assesment Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Taze dataframe doretmeli

```python
hours = np.round([2.2,5.7,4.4,3.6,0.9,0.9,0.3,5.2,3.6,4.2,0.1,5.8,5,1.3,1.1,1.1,1.8,3.1,2.6,1.7,3.7,0.8,1.8,2.2,2.7])
score = np.round([37.4,100,100,67.8,25.3,18.9,9.2,97.4,67,67.4,1.6,95.7,79.6,29.1,24.1,30.4,24.7,68.1,76.7,48.1,67.3,31.4,26.2,27.1,58.8])


dataframe = pd.DataFrame({'StudentID': range(234701, 234726),
                          'Study Hours': hours,
                          'Exam Score': score})
dataframe
```

```python
dataframe.describe()
```

```python
plt.scatter(dataframe['Study Hours'], dataframe['Exam Score'])
plt.xlabel('Study Hours (X)')
plt.ylabel('Exam Score (Y)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = dataframe[['Study Hours']]
y = dataframe['Exam Score']
model.fit(X,y)

intercept = model.intercept_
slope = model.coef_[0]

print(f"β0: {intercept:.2f}")
print(f"β1: {slope:.2f}")
print()

print(f"Regression Equation: Exam Score = {intercept:.2f} + {slope:.2f} * Study Hours")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Impact of Study Hours on Academic Performance')
plt.legend()
plt.show()
```

### Error-y tapmak

```python
rows = []
for i in dataframe.index:
    x = dataframe.loc[i, 'Study Hours']
    actual = dataframe.loc[i, 'Exam Score']
    pred = model.predict([[x]])[0]
    rows.append({
        'Study Hours': x,
        'Actual Exam Score': actual,
        'Predicted Exam Score': pred,
        'Residual': actual - pred
    })

new_df = pd.DataFrame(rows)
```
